```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
from openpyxl import Workbook, load_workbook
```

```python
import datetime

c_date = datetime.date.today() #current_date
#starting creating excel file
try:
    wb = load_workbook(f'cricket_yuv_t20_2007{c_date}.xlsx')
    active_sheet = wb['Sheet']
except FileNotFoundError:
    wb = Workbook()
    active_sheet = wb['Sheet']
    active_sheet.append(['Inns','Score','Overs','Conc','Wkts','Ct','St','vs','Opposition','Ground','Start DateAscending','dd'])
    wb.save(f'cricket_yuv_t20_2007{c_date}.xlsx')
except:
    print('something went wrong')



url='https://stats.espncricinfo.com/ci/engine/player/36084.html?class=1;spanmax1=05+Dec+2007;spanval1=span;template=results;type=allround;view=innings'
chrome_options = Options()
chrome_options.headless = False
browser = webdriver.Chrome(chrome_options=chrome_options,executable_path ='C:\\Users\\thaku\\Desktop\\inte\\chromedriver')
# browser = webdriver.Chrome(s)
browser.maximize_window()

browser.get(url)

browser.find_element_by_xpath('//*[@id="statsTab"]/li[3]/a').click()#to chane link

table = browser.find_element_by_xpath("//table[@class='engineTable']/caption[contains(text(),'Innings by innings list')]")

rows = len(table.find_elements_by_xpath("//tr[@class='data1']"))
print(rows)
columns =  len(table.find_elements_by_xpath("//tr[@class='data1']/td"))
print(columns)
for row in range(2,rows-1):
    data_list=[]

    
    for col in range(1,13):
        
        print()
        
        data_list.append(table.find_element_by_xpath(f"//tr[@class='data1']{[row]}/td{[col]}").text)
        
    if data_list not in active_sheet:
        active_sheet.append(data_list)
    

wb.save(f'cricket_yuv_t20_2007{c_date}.xlsx')
```
